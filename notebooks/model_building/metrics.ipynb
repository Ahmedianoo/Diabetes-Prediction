## Model Evaluation Metrics

The target of this project is to predict whether a patient is **Diabetic** or **Not Diabetic**.  
Therefore, this is a **binary classification problem**.

Before interpreting the evaluation metrics, it is important to understand the four outcomes of the confusion matrix.

| Term | Meaning in This Project |
|---|---|
| **True Positive (TP)** | The patient is actually diabetic, and the model correctly predicts **Diabetic**. |
| **True Negative (TN)** | The patient is actually not diabetic, and the model correctly predicts **Not Diabetic**. |
| **False Positive (FP)** | The patient is actually not diabetic, but the model incorrectly predicts **Diabetic**. |
| **False Negative (FN)** | The patient is actually diabetic, but the model incorrectly predicts **Not Diabetic**. |

In this problem, **False Negatives (FN)** are especially important because they mean that a diabetic patient was missed by the model.

---

## Accuracy

**Accuracy** measures the percentage of total predictions that the model classified correctly.

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

A high accuracy means the model is correct most of the time. However, accuracy can be misleading when the dataset is imbalanced.  
For example, if most patients are not diabetic, the model may achieve high accuracy by predicting most patients as **Not Diabetic**, while still missing many diabetic patients.

---

## Precision

**Precision** measures how many of the patients predicted as diabetic are actually diabetic.

$$
Precision = \frac{TP}{TP + FP}
$$

High precision means that when the model predicts **Diabetic**, it is usually correct.

In this project, precision helps reduce **False Positives**, meaning it reduces the number of non-diabetic patients incorrectly classified as diabetic.

---

## Recall

**Recall**, also called **Sensitivity**, measures how many actual diabetic patients were correctly detected by the model.

$$
Recall = \frac{TP}{TP + FN}
$$

High recall means the model is good at identifying diabetic patients.

In this project, recall is very important because a low recall means the model misses diabetic patients. Missing a diabetic patient can be more harmful than incorrectly warning a non-diabetic patient.

---

## F1-score

**F1-score** is the harmonic mean of precision and recall. It combines both metrics into one balanced score.

$$
F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

F1-score is useful when we need a balance between:

- Correctly predicting diabetic patients
- Detecting as many diabetic patients as possible

A high F1-score means the model has a good balance between precision and recall.

---

## ROC-AUC

**ROC-AUC** stands for **Receiver Operating Characteristic - Area Under the Curve**.

It measures the model's ability to distinguish between diabetic and non-diabetic patients across different classification thresholds.

A higher ROC-AUC means the model is better at separating the two classes.

| ROC-AUC Value | Interpretation |
|---:|---|
| 0.50 | Random performance |
| 0.70 - 0.80 | Acceptable |
| 0.80 - 0.90 | Good |
| 0.90 - 1.00 | Excellent |

ROC-AUC is useful for evaluating how well the model ranks patients by risk, but it does not directly show how many diabetic patients are missed at a specific threshold.

---

## PR-AUC

**PR-AUC** stands for **Precision-Recall Area Under the Curve**.

It measures the trade-off between precision and recall across different classification thresholds.

PR-AUC is especially useful when the dataset is imbalanced, meaning one class appears much more often than the other.

In this project, if diabetic patients are less frequent than non-diabetic patients, PR-AUC is more informative than ROC-AUC because it focuses on the positive class, which is **Diabetic**.

A higher PR-AUC means the model can detect diabetic patients while maintaining good precision.

---

## MCC

**MCC** stands for **Matthews Correlation Coefficient**.

It is a balanced evaluation metric that considers all four parts of the confusion matrix:

$$
TP,\ TN,\ FP,\ FN
$$

The MCC value ranges from **-1** to **1**.

| MCC Value | Interpretation |
|---:|---|
| 1 | Perfect prediction |
| 0 | Random prediction |
| -1 | Completely wrong prediction |

MCC is useful for imbalanced datasets because it gives a fair overall evaluation of the model's performance.

A high MCC means the model performs well on both diabetic and non-diabetic patients.

---

## Metric Priority in This Project

Since this project predicts whether a patient is diabetic or not, the most important error is a **False Negative**:

> The patient is actually diabetic, but the model predicts Not Diabetic.

This type of error is dangerous because the patient may not receive the needed medical attention.

Therefore, the most important metrics for this project are:

| Priority | Metric | Why It Matters |
|---:|---|---|
| 1 | **Recall** | Measures how many diabetic patients the model successfully detects. |
| 2 | **F1-score** | Balances precision and recall. |
| 3 | **PR-AUC** | Evaluates performance on the diabetic class, especially with imbalanced data. |
| 4 | **MCC** | Gives a balanced overall evaluation using TP, TN, FP, and FN. |
| 5 | **Accuracy** | Useful as a general metric, but should not be used alone. |

---

## Final Note

For this diabetes prediction task, the model should not only achieve high accuracy.  
It should also achieve strong **Recall**, **F1-score**, **PR-AUC**, and **MCC**.

A model with very high accuracy but low recall may still be risky because it can miss diabetic patients.
